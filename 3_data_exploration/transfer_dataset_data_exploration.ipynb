##ðŸ“Š Data Visualization

```python
# Top 10 Most Expensive Transfers
top_transfers = df.sort_values("Transfer Fee", ascending=False).head(10)
sns.barplot(data=top_transfers, x="Transfer Fee", y="Player Name", palette="viridis")
plt.title("Top 10 Most Expensive Transfers")
```

```python
# Total Transfer Spending by Season
season_spending = df.groupby("Season")["Transfer Fee"].sum().reset_index()
sns.lineplot(data=season_spending, x="Season", y="Transfer Fee", marker="o")
plt.title("Total Transfer Spending by Season")
```

```python
# Number of Transfers by Position
sns.countplot(x="Position", data=df, order=df["Position"].value_counts().index)
plt.title("Number of Transfers by Position")
plt.xticks(rotation=45, ha="right")
```

##ðŸ“Š Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")

# 1. Distribution of Average Rating (Pre vs Post Transfer)
plt.figure(figsize=(12, 5))
sns.histplot(
    pre["Average Rating"], kde=True, color="blue", label="Pre Transfer", stat="density"
)
sns.histplot(
    post["Average Rating"],
    kde=True,
    color="green",
    label="Post Transfer",
    stat="density",
)
plt.title("Distribution of Average Rating (Pre vs Post Transfer)")
plt.xlabel("Average Rating")
plt.legend()
plt.tight_layout()
plt.show()
```
